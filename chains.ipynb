Invoke Method

```python
from langchain_openai import AzureChatOpenAI,AzureOpenAIEmbeddings
from langchain.prompts import PromptTemplate
from dotenv import load_dotenv,find_dotenv
import os

load_dotenv(find_dotenv(),override=True)
endpoint = os.getenv("ENDPOINT_URL")
subscription_key = os.getenv("AZURE_OPENAI_API_KEY")
llm = AzureChatOpenAI(azure_endpoint=endpoint,api_key=subscription_key,api_version="2025-01-01-preview")

template = "give me 5 name of  {item}"
prompt_template = PromptTemplate(template=template, input_variables=["item"])

result = prompt_template.invoke({"item": "us"})
res=llm.invoke(result)
print(res)
```

Sequential Chain | RunnableSequence

```python
from langchain_openai import AzureChatOpenAI,AzureOpenAIEmbeddings
from langchain.prompts import PromptTemplate
from dotenv import load_dotenv,find_dotenv
import os
from langchain_core.runnables import RunnableSequence

load_dotenv(find_dotenv(),override=True)
endpoint = os.getenv("ENDPOINT_URL")
subscription_key = os.getenv("AZURE_OPENAI_API_KEY")
llm = AzureChatOpenAI(azure_endpoint=endpoint,api_key=subscription_key,api_version="2025-01-01-preview")

template = "give me 5 name of  {item}"
prompt_template = PromptTemplate(template=template, input_variables=["item"])

chain=RunnableSequence(prompt_template,llm)
result=chain.invoke({"item": "us"})

print(result.content)
chain.get_graph().print_ascii()
```
