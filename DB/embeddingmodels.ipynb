HuggingFaceEmbeddings

```python
from langchain_google_genai import ChatGoogleGenerativeAI,GoogleGenerativeAIEmbeddings
from langchain.prompts import PromptTemplate
from dotenv import load_dotenv,find_dotenv
import os
import numpy as np
from langchain_huggingface import HuggingFaceEmbeddings 

load_dotenv(find_dotenv(),override=True)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

res=embeddings.embed_query("how are you?ky ahaal chal hai bhao har har modi ")
np.shape(res)
```

GoogleGenerativeAIEmbeddings

```python
from langchain_google_genai import ChatGoogleGenerativeAI,GoogleGenerativeAIEmbeddings
from langchain.prompts import PromptTemplate
from dotenv import load_dotenv,find_dotenv
import os
import numpy as np

load_dotenv(find_dotenv(),override=True)
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

res=embeddings.embed_query("tell me capital of japan")
np.shape(res)
```

AzureOpenAIEmbeddings

```python
from langchain_openai import AzureChatOpenAI,AzureOpenAIEmbeddings
from langchain.prompts import PromptTemplate
from dotenv import load_dotenv,find_dotenv
import os
import numpy as np

load_dotenv(find_dotenv(),override=True)

embedding_endpoint = os.getenv("EMBEDDING_ENDPOINT_URL")
embedding_subscription_key = os.getenv("EMBEDDING_AZURE_OPENAI_API_KEY")
llm_embedding = AzureOpenAIEmbeddings(azure_endpoint=embedding_endpoint,api_key=embedding_subscription_key,model="text-embedding-ada-002")

res = llm_embedding.embed_documents("give me 5 state  of japan","ram ji ki jai ho")
np.shape(res[1])
```
