```python
from typing import TypedDict
from langgraph.graph import END,StateGraph

class InputState(TypedDict):
    string_value: str
    numeric_value: str

def modify_input_state(input_state: InputState) -> InputState:
    print(f"{input_state['string_value']}, {input_state['numeric_value']}")
    input_state['string_value'] = input_state['string_value'].upper()
    input_state['numeric_value'] = str(int(input_state['numeric_value']) + 1)
    return input_state

graph=StateGraph(InputState)
graph.add_node("branch_a", modify_input_state)
graph.add_node("branch_b", modify_input_state)
graph.add_edge("branch_a", "branch_b")
graph.add_edge("branch_b", END)
graph.set_entry_point("branch_a")

input_state: InputState = { "string_value": "hello","numeric_value": "1" }   

runnable = graph.compile()
runnable.invoke(input_state)

```
