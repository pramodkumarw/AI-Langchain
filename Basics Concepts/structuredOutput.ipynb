```python
from typing import TypedDict,Annotated
import os
from dotenv import load_dotenv,find_dotenv
from langchain_openai import AzureOpenAIEmbeddings,AzureChatOpenAI  

load_dotenv(find_dotenv(),override=True)
```

```python
class ProductReview(TypedDict): 
 Summary: Annotated[str , "A brief summary of the review"] 
 Sentiment: Annotated[str , "Return sentiment of the review either negative, positive or neutral"]


llm=AzureChatOpenAI(azure_endpoint=os.getenv("ENDPOINT_URL"),api_key=os.getenv("AZURE_OPENAI_API_KEY"),api_version=os.getenv("API_VERSION"))
structured_llm=llm.with_structured_output(ProductReview)
structured_llm.invoke("I’ve been using these XYZ Wireless Earbuds for two weeks, and I’m impressed! The sound quality is deep and clear, and they are lightweight enough for long gym sessions")
```
